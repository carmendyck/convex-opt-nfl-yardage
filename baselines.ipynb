```python
import numpy as np
```

```python
#load data and define variables

P = np.load('dataset/probability_data_min_max_normed.npy') #passing_prob
N = 27 #number_players
v = np.load('dataset/player_ids.npy') #player_ids
y = np.load('dataset/passing_yards.npy') #passing yards
z = np.load('dataset/receiving_yards.npy') #rushing yards
```

```python
#change these variables

a = np.random.randint(1, 10, size=P.shape[0])#cost matrix
A = 100 #budget
w = np.full(P.shape[0], 10)
```

```python
# Greedy approach: Select top N players while staying within budget

player_wise_performance = y+z
performance_dict = {}
for idx,value in enumerate(player_wise_performance):
  performance_dict[idx] = value

sorted_performance = sorted(performance_dict.items(), key=lambda x: x[1], reverse=True)

selected_players = []
total_cost = 0
total_performance = 0

for player, performance in sorted_performance:
    cost = a[player]
    if total_cost + cost <= A and len(selected_players) < N:
        selected_players.append(player)
        total_cost += cost
        total_performance += performance

print("Selected Players:")
for player in selected_players:
    print(f"Player ID: {v[player]}, Performance: {player_wise_performance[player]}")

print("Total Cost:", total_cost)
print("Total yardage/performance:", total_performance)
```

cons- with low budget won't reach N players


```python
# Heuristic approach: Select top N players with the best performance/cost ratio while staying within budget

player_data = [(idx, performance, cost) for idx, (performance, cost) in enumerate(zip(player_wise_performance, a))]
sorted_players = sorted(player_data, key=lambda x: x[1] / x[2], reverse=True)

selected_players = []
total_cost = 0
total_performance = 0

for player, performance, cost in sorted_players:
    if total_cost + cost <= A and len(selected_players) < N:
        selected_players.append((player, performance))
        total_cost += cost
        total_performance += performance

print("Selected Players:")
for player, performance in selected_players:
    print(f"Player ID: {v[player]}, Performance: {performance}")

print("Total Cost:", total_cost)
print("Total yardage/performace:", total_performance)
```

cons - might not utilize budget fully, this is only selecting cost efficient players, will select some low cost weak players
pros - unless budget is very low, this approach will give full N player roster
