```python
import numpy as np
```

```python
#load data and define variables

P = np.load('dataset/probability_data_min_max_normed.npy') #passing_prob
N = 27 #number_players
v = np.load('dataset/player_ids.npy') #player_ids
y = np.load('dataset/passing_yards.npy') #passing yards
z = np.load('dataset/receiving_yards.npy') #rushing yards
```

```python
#change these variables

a = np.random.randint(1, 10, size=P.shape[0])#cost matrix
A = 100 #budget
w = np.full(P.shape[0], 0.1)
```

```python
#solve Primal

import cvxpy as cp

M = len(v)
x = cp.Variable(M, boolean=True)

objective = cp.Maximize(cp.sum(cp.multiply(y.T @ P, x)) + cp.sum(cp.multiply(z.T @ P, x)))


constraints = [a @ x <= A,
               cp.sum(x) == N]


problem = cp.Problem(objective, constraints)

problem.solve()

optimal_x = x.value

print("Optimal team selection vector:")

for i, val in enumerate(optimal_x):
    if val == 1:
        print(f"Player ID: {v[i]}")

print("Total yardage available by the team:", objective.value)
```

```python
#solve Dual

M = len(a)
lambda_var = cp.Variable()
nu = cp.Variable()

objective = cp.Maximize(-lambda_var*A - nu * N)

constraints = [lambda_var * a + nu * np.ones(M) - y.T @ P - z.T @ P == 0,
               lambda_var >= 0]


problem = cp.Problem(objective, constraints)


problem.solve()

optimal_lambda = lambda_var.value
optimal_nu = nu.value

print("Optimal values of lambda:", optimal_lambda)
print("Optimal value of nu:", optimal_nu)
```
